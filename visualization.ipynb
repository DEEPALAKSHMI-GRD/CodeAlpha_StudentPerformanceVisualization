```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\deepa\Downloads\studentperformance.csv")
df.head()
```

```python
sns.countplot(x="FinalResult", data=df)
plt.title("Final Result Distribution")
plt.show()
```

```python
sns.countplot(x="Gender", hue="FinalResult", data=df)
plt.title("Gender vs Final Result")
plt.show()
```

```python
sns.boxplot(x="FinalResult", y="StudyHours", data=df)
plt.title("Study Hours vs Final Result")
plt.show()
```

```python
sns.histplot(df["Attendance"], kde=True)
plt.title("Attendance Distribution")
plt.show()
```

```python
df[["MathScore","ScienceScore","EnglishScore"]].boxplot()
plt.title("Subject Score Distribution")
plt.show()
```

```python
df["AverageScore"] = df[["MathScore","ScienceScore","EnglishScore"]].mean(axis=1)

sns.scatterplot(x="StudyHours", y="AverageScore", hue="FinalResult", data=df)
plt.title("Study Hours vs Average Score")
plt.show()
```

```python
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
